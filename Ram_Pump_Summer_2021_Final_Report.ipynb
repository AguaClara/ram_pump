## Introduction

In AguaClara plants, the flow of water throughout the plant is determined solely by gravity; all treated water exits at the lowest part of the plant. Currently, plant operators are required to manually fill buckets of treated water from the bottom of the plant and carry them to the top, where they use this water to continuously fill chemical stock tanks. This work can be cumbersome for plant workers; however, hydraulic ram pumps seem to be a plausible solution to this challenge. 

The purpose of a hydraulic ram pump is to utilize the kinetic energy of falling water in order to pump the water back upwards. In AguaClara plants specifically, the untreated water will make its way through the plant, falling to a lower elevation with each step. Once the water gets through the rapid sand filters (the last step of water treatment), it will reach the ram pump, where some treated water will continue downwards to the distribution system, while the rest will be pumped to the chemical stock tanks at the top of the plant (Figure 1). The current elevation between where the ram pump is installed to the surface of the filtered water channel is approximately 2 meters, and the elevation from the surface to the top of the chemical stock tank is approximately 4 meters. Thus, the AguaClara Vertical Ram Pump (ACVRP) needs to be able to pump water to a minimum of 6 meters of head. Ideally, there will be a constant flow of treated water via the ram pump in order to maintain the necessary concentration of solvent in the chemical stock tanks.

## Literature Review and Previous Work

### Ram Pump Design 

Conventional ram pumps are designed horizontally in order to pump water from a lower elevation to a higher elevation. The head tank is placed at a higher elevation to provide potential energy that ultimately powers the pump. The drive pipe directs water to the pump, where there is a valve that can be opened and closed (via a spring system). Once the valve is closed, water is redirected and stored in an air chamber that is already pressurized. The air pressure then pushes the water upwards to where it is needed next. The AguaClara Vertical Ram Pump (ACVRP) bears many similarities to this conventional design. However, changes were implemented by previous teams in order to minimize the cost in plants and maximize their efficiency given the already existing technology/infrastructure in the AguaClara plants. 

The Summer 2019 team, consisting of members Ching Pang and Alyssa Ju [(Pang and Ju, 2019)](https://github.com/AguaClara/ram_pump/blob/master/Summer%202019/RamPump2019Summer.ipynb), focused heavily on redesigning certain aspects of the Ram Pump hardware in order to decrease head loss and observe more consistent pressure cycles. They designed the new ram pump system to lie along a vertical axis, thus making it easier to incorporate in the existing piping. In addition, they increased the length of the air chamber, increased the effluent valve diameter, and included unions that maintained consistent dimensions after each assembly (Figure 2). Experimentation confirmed that increasing the diameter of the effluent valve decreased the amount of headloss of the water when it flowed into the effluent valve. Lastly, the team implemented a new offset t-joint for the air chamber. This design led to the most consistent pressure cycles. In particular, the team found that it was easier to maintain a constant pressure within the air chamber when water was already initially accumulated to the height of the t-joint (Figure 3). ### Spring System 

The spring system is of vital importance within the ram pump. This system determines whether the valve is open, in which case the water goes straight to the distribution system, or closed, when the water is redirected into the air chamber and pumped up vertically. Essentially, the spring system relies heavily on forces coming from the water and the spring itself. In the initial state, the plate is open; the water is able to continue downward past the effluent valve and into the distribution system (or waste bucket in the experimental setup). As the water flows downward, it begins to accelerate. This acceleration creates a viscous drag force that is able to overcome the spring force, which then closes the plate. The water is redirected through the effluent valve and into the air chamber. The pressure difference pumps the water upward. As the water is flowing into the effluent valve, there is considerable headloss, which decreases the velocity of water, in turn decreasing the drag force. The decreased drag force allows the spring force to dominate and reopen the plate, which causes the water to flow once again into the waste bucket. This alternating cycle between closed and open states is the mechanism ultimately responsible for driving the water upward.

In addition, the Fall 2019 team made considerable progress towards a new design for the spring system as a whole. The old spring design consisted of one spring located within the check valve at the top of the ram pump (Figure 5). The team came across several challenges with this design. First, there was no way to independently adjust the forces for when the plate is open versus when the plate is closed. Since there was only one spring with a stopper nut to be adjusted, the adjustment would affect both closing and opening the valve. Secondly, in order to make such an adjustment, the entire ram pump would have to be disassembled, since the spring was located inside the valve pipe. In order to combat these challenges, the team designed a new three spring system.

The new spring system consists of three springs inline attached to two plates. On the top, the springs are connected to the plate via eyebolts. The two outer springs are of the same length and spring constant, while the middle is shorter and is attached to a rope with slack when the plate is open. The bottom plate attaches to all three springs and also a metal rope that connects all the way to the plate that opens and closes, located in the check valve. See Figure 6 for a detailed representation. This new design addresses the challenges with the previous iteration. Because of the slack of the rope attached to the middle spring, only the two outer springs contribute to the spring force acting on the plate when it is open. Meanwhile, all three springs act on the plate when it is closed. Thus, the forces act independently and can be adjusted separately in order to maximize the force on either the open or closed plate. Secondly, placing the spring system above the ram pump and head tank allows the springs and thus their forces to be adjusted instantaneously throughout the experiment; no disassembly is required.

### Experimental Results 

Previous AguaClara teams have come up with several distinct experimental procedures, most of them rooted in first setting up the apparatus correctly and tweaking certain aspects of the design. As discussed previously, both the air chamber, effluent valve, and spring system were adjusted in order to increase the ram pump efficiency. 

In order to test the new air chamber design, the summer 2019 team calculated the headloss of the water when it flowed through the effluent valve, and found it had decreased to 0.62 percent of the previous design. Moreover, the efficiency of the ram pump was increased from a mere 8 percent to over 40 consistently. The efficiency was calculated by comparing theoretical flow rates with experimental flow rates. The same team also found more consistent flow rates when the optimal springs were used. A consistent flow rate is important for the chemical stock tanks to hold the right amount of coagulant [(Pang and Ju, 2019)](https://github.com/AguaClara/ram_pump/blob/master/Summer%202019/RamPump2019Summer.ipynb). 

The fall 2019 team tested their new spring apparatus in order to determine its effectiveness. The goal was to adjust the middle spring and determine its effect on the flow rate out of the air chamber. The team started with the eyebolt all the way down, and would subsequently turn the eyebolt until it began to cycle. Next, they let the ram pump run for about a minute until turning the bolt one more full turn. The process continued until the ram pump no longer cycled. The team completed this experiment with three different springs. The team found that the average flow rate was over 20 mL/s, and the highest flow rate achieved was 26.735 mL/s. This was a significant improvement from the previous data collected, which achieved an average flow rate of less than 20 mL/s, keeping in mind that their setup had a larger volume of water in the head tank. Thus, the team was able to conclude that their spring model was more efficient. The analysis of the flow rate gave insight into the pumping power of the overall model. Instead of focusing on the spring constants itself, the team wanted to find the best set of springs for all of the different plant sizes as a whole [(Pang, Storch and Hunter, 2019)](https://github.com/AguaClara/ram_pump/blob/master/Fall%202019/RamPump2019Fall.ipynb). 


The Spring 2021 team focused their attention on finding ways to determine and decrease the uncertainty in parts of the experimental setup pertaining to the flow rate calculations. They identified a couple key areas as the major sources of uncertainty: turbulence in the head tank, spring system adjustments, and sealing the air chamber. While the team could not physically test these ideas in the lab, they did come up with some viable solutions. They proposed implementing a potential crank system that could turn the eyebolts mechanically and more accurately. In addition, in order to combat the turbulence in the head tank, they suggested increasing the diameter of the water input pipe, or adding another pipe to more evenly distribute the water throughout the head tank [(Pang, Chen and Hunter, 2021)](https://github.com/AguaClara/ram_pump/blob/master/Report_Spring2021.ipynb).

### **Processing Data using Spring Constants**
To convert the location of the clamps for the weak spring to usable data, their relative positions were converted into the amount of the spring's stretch when the check valve closes with each cycle. Then, this value was converted into forces in Newtons.

To convert the measured values into the spring stretch, the amount of stretch at the previously determined zero point was measured. Then, the measurement in centimeters was added onto the values labeled "Weak (cm)".

The forces were calculated by using Hooke's law, $F = -kx$. To determine $k$, the spring constant, various objects of known weight were attached to the springs and the spring's change in length was measured. The data from multiple weights were plotted to create a linear regression, the slope of which is equivalent to $-k$. The weak spring at equilibrium is 7.6 cm, and the spring used to determine the spring constant is identical on all aspects except its length, 15.35 cm. From the experiment, the 15.35 cm spring was determined to have a spring constant of 29.9 N/m. The following equation was used to determine the spring constant of the spring used in the system: $29.9*15.35/7.6=60.4$ $N/m$.

## Theoretical Calculations
To analyze the efficiency of the ram pump, the theoretical performance of the ram pump was modeled. To do this, the terminal velocity of water through the drive pipe was experimentally measured to be 1.603 m/s. Then, using the following formula, the velocities of the water in the drive pipe when the check valve opens and closes were modeled:

<center>
Acceleration: $v(t)=v_t*tanh(g/v_t*t)$
<br />
Deceleration: $v(t)=-g/L(z_3-z_1)*(t-t_i)+v_i$
</center>

Where $v_t$ is the terminal velocity (1.603 m/s), $L$ is the length of the drive pipe (1.56 m), $z_3$ is the height of pumped water (6 m), and $z_1$ is the height of the water in the head tank (1.56 m). $t_i$ and $v_i$ are the time and velocity at the point when the check valve closes. Using code, various cycle times were experimented to see the response of the flow rate. The cycle times were determined by adjusting the fraction of the terminal velocities at which the valve closes (x) and opens (y).

</center>

## Conclusions
The maximum experimental flow rate was found to be 26.55 mL/s, therefore the efficiency in relation to the absolute maximum theoretical flow rate of 89.57 mL/s was 29.6%. However, this value is not useful in measuring the potential of the ram pump system. Because the theoretical model does not account for the non-zero period of time required to open and close the valve, a flow rate that results from an impossibly short cycle time should not be used to compare with the experimental results. A more reasonable cycle time with $y=0$ in Figure 22 would be a better measure of performance because the experimental system matches more closely with the check valve's opening and closing points. With the theoretical flow rate of 57.87 mL/s, the system's efficiency was adjusted to 45.9%.

This measure of performance allows for future adjustments to be better quantified and understood because each adjustment of springs can be interpreted in terms of points at which the check valve closes and opens. Understanding this relation will help determine whether an improvement in design is due to changing these points or a potential undiscovered limitation with the current design.

# Manual


## Experimental Methods
### Set-up
* Fill the waste bucket with water until the sump pump is covered at least half-way.
* Ensure that the RJ45 I/O connections are properly installed to the ProCoDA box.
* Install the ProCoDA method file and configure the PID using [this procedure]( https://monroews.github.io/EnvEngLabTextbook/ProCoDA/ProCoDA.html#proportional-integral-derivative-control).

### Experiment
* Select the file location for short-term experiments and set the data interval to 0.1 seconds.
* Change the state to "Peristaltic On".
* Ensure that the ram pump is operational and the water level in the head tank is at the height of the drain pipe.
* Wait until the peristaltic pump speed reaches equilibrium.

### Cleaning Procedure
* Drain the air chamber by changing the state to "Peristaltic Only".

```
import aguaclara as ac
from aguaclara.core.units import u
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import scipy.optimize as opt
from aguaclara.play import *
```

```
g = 9.81
H = 1.56
L = H
z1 = L
z3 = 6
v = 1.603

def plot(start, end):
  # Adjust start and end points
  start_pt = start
  end_pt = end
  start_v = start_pt * v
  end_v = end_pt * v
  t = np.arange(0, 0.5, 0.01)
  deceleration = -g/L*(z3-z1)

  def acc(t):
    return v * np.tanh(t*g/v)

  def acc_approx(t):
    return g/L*z1*t

  def dec(t):
    return deceleration*(t - start_t) + start_v

  def inv_acc(a):
    return v/g*np.arctanh(a/v)

  def inv_dec(d):
    return (d - start_v) / deceleration + start_t

  start_t = inv_acc(start_v)
  end_t = inv_dec(end_v)
  start_a_t = inv_acc(end_v)
  cycle_time = end_t - start_a_t
  answer, error = sp.integrate.quad(dec, start_t, end_t)
  def vol(x):
    return x * math.pi * (0.5 ** 2) / 1550 * 1000000  # mL/cycle
  Q = vol(answer)/cycle_time
  print(str(start_pt*100) + "%, " + str(end_pt*100) + "%: " + str(vol(answer)) + " mL/cycle, " + str(Q) + " mL/s") 

  # test_x = 0.5 * L/g * v**2 * (start_pt**2 - end_pt**2)/(z3-z1)
  # test_Q = vol(test_x) / (v/g*(np.arctanh((start_pt-end_pt)/(1-start_pt*end_pt))+L*(start_pt-end_pt)/(z3-z1)))
  # print(vol(test_x))
  # print(test_Q)

  plt.title("Theoretical Model of the Ram Pump")
  plt.xlabel("time (sec)")
  plt.ylabel("velocity (m/s)")
  plt.ylim(0,2)
  plt.plot(t,acc(t), 'r')
  # plt.plot(t,acc_approx(t), 'm')
  plt.plot(t,dec(t), 'g')

  # Previous and next cycles (for demonstration)
  plt.plot(t,dec(t+cycle_time),'g:')
  plt.plot(t,acc(t-cycle_time), 'r:')
  plt.plot(t,dec(t-cycle_time), 'g:')
  plt.plot(t,acc(t-2*cycle_time), 'r:')

  # Points
  plt.plot(start_a_t, end_v, 'bo')
  plt.plot(start_t, start_v, 'bo')
  plt.plot(end_t, end_v, 'bo')
  plt.plot(start_t + cycle_time, start_v, 'bo')
  plt.plot(end_t + cycle_time, end_v, 'bo')

  plt.show()
```
